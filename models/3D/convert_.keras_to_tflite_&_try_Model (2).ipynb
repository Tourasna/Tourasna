```python
import tensorflow as tf

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÖŸàÿØŸäŸÑ
model = tf.keras.models.load_model("landmark_classifier.keras")

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸÖŸàÿØŸäŸÑ ŸÑÿµŸäÿ∫ÿ© TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# ÿ≠ŸÅÿ∏ ÿßŸÑŸÖŸÑŸÅ ÿßŸÑŸÜÿßÿ™ÿ¨
with open("landmark_classifier.tflite", "wb") as f:
    f.write(tflite_model)

print("‚úÖ ÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ° landmark_classifier.tflite ÿ®ŸÜÿ¨ÿßÿ≠!")
```

```python
import numpy as np
import tensorflow as tf
from PIL import Image
import json

# Load TFLite model
interpreter = tf.lite.Interpreter(model_path="landmark_classifier.tflite")
interpreter.allocate_tensors()

# Load labels from JSON list
with open("label_map.json", "r") as f:
    labels = json.load(f)  # labels is now a list

# Get input/output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Load image
img_path = "/content/WhatsApp Image 2025-10-13 at 8.17.11 PM.jpeg"  # ÿ∫Ÿäÿ±Ÿá ŸÑŸÖÿ≥ÿßÿ± ÿßŸÑÿµŸàÿ±ÿ© ÿπŸÜÿØŸÉ
img = Image.open(img_path).convert('RGB')

# Resize ÿßŸÑÿµŸàÿ±ÿ© ŸÑÿ™ÿ™ŸÜÿßÿ≥ÿ® ŸÖÿπ ÿßŸÑŸÖŸàÿØŸäŸÑ
input_shape = input_details[0]['shape']
height, width = input_shape[1], input_shape[2]
img = img.resize((width, height))

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿµŸàÿ±ÿ© ŸÑŸÖÿµŸÅŸàŸÅÿ© numpy
input_data = np.array(img, dtype=np.float32)

# Normalize (0-1)
input_data = input_data / 255.0

# Add batch dimension
input_data = np.expand_dims(input_data, axis=0)

# Run inference
interpreter.set_tensor(input_details[0]['index'], input_data)
interpreter.invoke()

# Get output
output_data = interpreter.get_tensor(output_details[0]['index'])

# Get predicted class
pred_idx = np.argmax(output_data)
pred_label = labels[pred_idx]

print("Predicted landmark:", pred_label)
```
